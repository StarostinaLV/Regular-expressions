**Задание 1**

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:

car_id = 'A222BC96’
Результат: Номер A222BС валиден. Регион: 96

car_id = 'АБ22ВВ193’
Результат: Номер не валиден

```python
# Импорт модуля re:
import re
```

```python
# Исходные данные:
#car_id = 'A222BC96'
#car_id = 'АБ22ВВ193'
car_id = 'КМ222ВС96'
```

```python
def car_number(car_id):
  # Согласно ГОСТ 50577-2018, п. 3.7, на рег.знаках могут применяться буквы А, В, Е, К, М, Н, О, Р, С, Т, У, Х (все они пишутся аналогично некоторым буквам латинского алфавита) и D.
  # Шаблон для проверки транспортного номера t_n (предусмотрен вариант, когда буквы номера введены на латиннице):
  t_n = re.findall(r'^[АВЕКМНОРСТУХABDEKMHOPCTYX]{1}\d{3}[АВЕКМНОРСТУХABDEKMHOPCTYX]{2}\d{2,3}', car_id)
  """
  Логика regex следующая:
  ^[АВЕКМНОРСТУХABDEKMHOPCTYX]{1} - поиск одной из перечисленных букв, причем до цифр буква должна быть только одна (поэтому ^),
  \d{3} - поиск следующих за первой буквой трех цифр,
  [АВЕКМНОРСТУХABDEKMHOPCTYX]{2} - за тремя цифрами следуют две разрешенные буквы,
  \d{2,3} - номер ТС завершается номером региона, который может состоять из двух или трех цифр.

  """
  # Шаблон региона (для использования при выводе результатов проверки):
  region = re.findall(r'\d{2,3}$', car_id)
  if t_n:
    good = f"Результат: номер {car_id} валиден. Регион: {region[0]}."
    return good
  else:
    return 'Результат: номер не валиден.'
```

```python
car_number(car_id)
```

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:

some_string = ‘Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.’

Результат:
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
# Исходные данные:
#some_string = """Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений."""
some_string = """Good good my happy happy guy! Oh just happy  happy  happy guy. Oh just a look at a happy, oh the happy dog, oh just a happy happy happy dog..."""
```

```python
# Искомая функция:
def reps_del(some_string):
  """
  Логика regex:
  (.\w+.) - до и после повторяющегося слова может стоять любой символ (на всякий случай). Символы объединены в группу, чтобы после сослаться на данную группу и заменить
  ее в исходном тексте ей же, но в одном экземпляре.
  Флаг re.I предусмотрен для случаев, когда повторяющееся слово написанно и с прописной, и со строчной буквы.

  """
  return re.sub(r'(.\w+.)\1+', r'\1', some_string, flags = re.I)
```

```python
reps_del(some_string)
```

**Задание 3**

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:

some_words = 'Информационные технологии’
Результат: ИТ

some_words = 'Near Field Communication’
Результат: NFC

```python
# Исходные данные:
#some_words = 'Информационные технологии'
some_words = 'Near Field Communication'
#some_words = 'Автоматика промышленных установок и технологических комплексов'
```

```python
def acronym(some_words):
  # Шаблон для поиска первых букв каждого слова:
  """
  Логика regex:
  ^\w{1} - поиск одной первой буквы строки,
  | - или (если это не первая буква строки),
  (?<=\s)[a-zа-яё] - поиск пробельного символа (без включения его в результат), за которым следует буква из указанного диапазона (которая включается в результат).
  Флаг re.I предусматривает случай, если часть слов написана с прописной буквы.

  """
  temp = re.findall(r'^\w{1}|(?<=\s)[a-zа-яё]', some_words, flags = re.I)
  # Преобразуем список полученных букв в строку, сделаем все буквы прописными.
  res = ''.join(temp).upper()
  # По условию задачи функция должна возвращать акроним, поэтому ниже применена инструкция return:
  return res
```

```python
acronym(some_words)
```

**Задание 4**

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:

emails = [‘test@gmail.com’, ‘xyz@test.in’, ‘test@ya.ru’, ‘xyz@mail.ru’, ‘xyz@ya.ru’, ‘xyz@gmail.com’]

Результат:

gmail.com: 2
test.in: 1
ya.ru: 2
mail.ru: 1

```python
# Исходные данные:
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
```

```python
def domen(emails):
  # Выводить результат будем в виде пар "ключ: значение", т.е. как элементы словаря.
  domen_dict = {}
  for email in emails:
    """
    Логика regex следующая:
    (?<=@)\w+ - поиск нескольких букв, цифр или символов _, которые следут за символом @ (при этом @ в результат не включается),
    \. - экранированная точка,
    \w+ - часть домена после точки, которая может состоять из нескольких символов.

    """
    domen_zone = re.search(r'(?<=@)\w+\.\w+', email).group()
    if domen_zone not in domen_dict:
      domen_dict[domen_zone] = 1
    else:
      domen_dict[domen_zone] += 1
  for key, value in domen_dict.items():
    print(key, value, sep = ': ')
```

```python
domen(emails)
```

**Задание 5 (необязательное)**

Напишите функцию, которая будет подсчитывать сколько слов начинается на гласные, а сколько на согласные буквы в тексте (текст может быть написан как с использованием букв кириллицы, так и латиницы).

Пример работы программы:

some_text = ‘Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.’

Результат:

Слов на гласные буквы: 9
Слов на согласные буквы: 21

```python
# Исходные данные:
some_text = """Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные."""
```

```python
def sounds_count(some_text):
# Шаблон для поиска начальных букв слов:
  temp = re.findall(r'^\w{1}|(?<=\s)[a-zа-яё]', some_text, flags = re.I)
  """
  Логика regex следующая:
  ^\w{1} - поиск в начале строки одной буквы,
  | - или,
  (?<=\s) - поиск пробельного символа, который в результат не включается,
  [a-zа-яё] - после пробельного символа производится поиск одной из букв в указанных диапазонах на кириллице или латиннице,
  регистр нижний, т.к. в шаблоне temp предусмотрен флаг, игнорирующий регистр.

  """
# Согласных больше, поэтому искать будем гласные:
  vowel_list = ['а', 'я', 'у', 'ю', 'о', 'е', 'ё', 'э', 'и', 'ы', 'a', 'e', 'i', 'o', 'u', 'y']
  vowel = 0
  consonant = 0
  for el in temp:
# Метод lower() - чтобы не "потерять" прописные начальные буквы.
    if el.lower() in vowel_list:
      vowel += 1
    else:
      consonant += 1
  print('Слов на согласные буквы:', consonant)
  print('Слов на гласные буквы:', vowel)
```

```python
sounds_count(some_text)
```
